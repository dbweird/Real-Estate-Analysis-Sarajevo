### Import libraries

```python
import pandas as pd
import numpy as np
```

### Reading scraped data

```python
data=pd.read_csv('real_estate_data_sarajevo.csv')
```

```python
data
```

### Getting relevant data (only "Vrsta oglasa"="Prodaja")

```python
data['Vrsta oglasa'].unique()
```

```python
data=data[data['Vrsta oglasa']=='Prodaja']
data
```

### Creating a new dataframe with relevant columns

```python
relevant_data=data[['id','Cijena','Stanje','Lokacija','Adresa','Kvadrata','Broj soba','Sprat','Vrsta grijanja',
                    'Vrsta poda','OLX ID','Broj pregleda','Godina izgradnje','Balkon','Kvadratura balkona','Lift',
                    'Parking','Garaža','Klima','Novogradnja','Alarm']].copy()
relevant_data
```

### Cleaning columns

Replacing NaN with "Nije uneseno"

```python
relevant_data = relevant_data.replace(np.nan, 'Nije uneseno', regex=True)
```

Cijena, Kvadrata, Kvadratura balkona

```python
relevant_data['Cijena'] = relevant_data['Cijena'].str.replace(r'KM', '')
relevant_data['Kvadrata']=relevant_data['Kvadrata'].str.replace(r',','.')
relevant_data['Kvadratura balkona']=relevant_data['Kvadratura balkona'].str.replace(r'j','.')
```

```python
relevant_data['Kvadrata']=relevant_data['Kvadrata'] = pd.to_numeric(relevant_data['Kvadrata'], errors='coerce')
```

```python
relevant_data=relevant_data.dropna(subset=['Kvadrata'])
```

Adresa

```python
relevant_data=relevant_data.replace({'Adresa' : { '12. mart' : '12. Mart', 'Željezničarska' : 'Željeznička',
                                                  'Žrtava fašizma' : 'Žrtava Fašizma',
                                                  'Adema B.':'Adema Buće','Adema Buce':'Adema Buće',
                                                  '8 Mart':'8. Mart','8.mart':'8. Mart','8. Marta':'8. Mart'}})
relevant_data['Adresa']
```

Broj soba

```python
relevant_data=relevant_data.replace({'Broj soba' : {'Četverosoban (4)':4,'Trosoban (3)':3,'Dvosoban (2)':2,'Jednosoban (1)':1,
                                                    'Jednoiposoban (1.5)':1,'Petosoban i više':5,'8+':8,'Garsonjera':1,
                                                    'Nije uneseno':0}})
```

```python
relevant_data['Broj soba']=relevant_data['Broj soba'].astype(int)
relevant_data['Broj soba'].value_counts()
```

Sprat

```python
relevant_data['Sprat']=relevant_data['Sprat'].replace('6+','6')
relevant_data['Sprat'].value_counts()
```

Godina izgradnje

```python
relevant_data=relevant_data.replace({'Godina izgradnje':{'2010+':'2010-2020','2015+':'2010-2019','Novogradnja':'2010-2019'}})
```

```python
relevant_data=relevant_data.replace("True",'Da')
```

```python
relevant_data
```

### Saving to csv

Saving data to .csv file

```python
relevant_data.to_csv('real_estate_data_SA.csv',index=False)
```
